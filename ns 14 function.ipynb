```python
def battery_low_warning():
  print('Battery Low!')
```

```python
battery_low_warning()
```

```python
def battery_manager(battery):
  if battery ==100:
    print('Battery Full!')
  elif battery >= 50 :
    print('Battery medium charge')  
  elif battery < 50 and battery >10:
    battery_low_warning()  #üstteki fonkisyonu yazdım buraya 
  else:
    print('Please charge your battery!')  
```

```python
battery_manager(35)
```

```python
battery_manager(70)
```

```python
battery_manager(5)
```

```python
def add_2(arg1):
  return arg1 + 2 # gonna store the output in the function(as variable)
```

```python
def add_3(x):
  return add_2(x) + 1   # burada x e yazdığın sayıyı ilk önce add_2 nin içine koymuş 
                        # oluyorsun o yüzden add_2 ye dönüp oradaki işlemi bitir sonra o sonucu add3 e koy
```

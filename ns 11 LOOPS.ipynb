```python
answer = 35

print('Let\'s play the guess a number game')

while True:
  guess = input('what a two digit number I have in mind? ')
  if not guess.isdigit():
      print('Please enter a numeric value.')
      guess =int(input('Please enter an integer: '))
  else:
      guess = int(guess)   
      if guess < answer:
          print('Go a little higher')
      elif guess > answer:
          print('Go little lower')
      else:
          print('What a mindreader!!')
      break    
```

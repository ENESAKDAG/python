```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day = input('Please enter a day:  ').strip().title()

if (day in days) and (day =='Saturday' or 'Sunday'):
    print(f'{day} is weekend')
elif (day in days) and (day != 'Saturday' and day != 'Sunday'):
    print(f'{day} is a weekend')
else:
    print(f'you have entered {day}.\nThis is not a valid day')
```

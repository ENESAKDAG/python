girilen sayının asal sayı olup olmadığını bulunuz

```python
num1 = int(input('Please enter a number: '))
count = 0

for i in range(1,num1+1):
   if num1%i== 0:
    count+=1

if num1==0 or num1 ==1 or count>2:
  print(f'Your number {num1} is not a prime number')
else:
  print(f'Your number {num1} is a prime number') 
```

 Klavyeden girilen sayıya kadar olan sayılardan tek sayıların toplamını ve çift sayıların toplamını ayrı ayrı bulan python kodunu yazınız.
 

```python
num1 = int(input('bir sayı giriniz:  '))
total_odd = 0
total_even = 0
for i in range(1,num1+1):
  if i%2 == 0:
    total_odd += i
  elif i%2 != 0: #buraya else girsen daha kısa sürer else den sonra hic bir sey yazmasanda olur.
    total_even += i

print(f'total odd :{total_odd} and total even :{total_even}')    

```

NESTED DICTIONARIES

```python
school_records={ # NESTED dict.
  'personal_info': # key
    {'kid':
     {'tom':{'class':'intermediate', 'age':10},
      'sue':{'class':'elemantary', 'age':8}
      },# bu kid in parantezi
    'teen':
     {'joseph':{'class':'college', 'age':19},
      'marry':{'class':'high school', 'age':16}
      },# BU  teen in parantezi
    }, # bu personal info  nun parantezi
  'grades_info': 
    {'kid':{'tom':{'math':88, 'speech':69}, 
      'sue':{'math':90, 'speech':81}
      },
    'teen':{'joseph':{'coding':80, 'math':89},
      'marry':{'coding':70, 'math':96}
      }, # teen in parantezi
    } #grades info nun parantezi
} #school records un parantezi
print(school_records['personal_info']['teen']['marry']['age']) 
print(school_records['grades_info']['teen']['marry']['coding']) 
```

```python
coding_grade = school_records['grades_info']['teen']['marry']['coding']
print(f'Marry\'s coding grade is : {coding_grade}')
```

```python
age_grade = school_records['personal_info']['teen']['marry']['age']
print(f'Marry\'s age_grade is : {age_grade}')
```

```python
sentence = input('Please enter a sentence:  ')
words = sentence.split()               
i = 0 #1 -->2 -->3 --> 4 --> 5 -->6                                   
longest_len = 0   # 3 -->7                             
while i < len(words):
  if len(words[i]) > longest_len:
     longest_len = len(words[i]) 
     longest = words[i]  # The --> Fastest
  i += 1
print(words)
print('the length of the longest word : ', longest_len)  
print('The length of the longest word in your sentence is ', longest) 
```

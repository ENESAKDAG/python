```python
from datetime import date as d 

birth = d(571, 4, 22)
death = d(632, 6, 8)

life_in_days = d.toordinal(death) - d.toordinal(birth)
life_in_years = life_in_days/365.25
print(life_in_days)
print(life_in_years)
```

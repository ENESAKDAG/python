Klavyeden girilen başlangıç ve bitiş sayıları arasında bulunan çift sayıların ortalamasını bulan python kodunu yazınız.

```python
num1 = int(input('başlangıç sayısını giriniz: '))
num2 = int(input('bitiş sayısını giriniz: '))
count = 0 
toplam = 0
for i in range(num1+1,num2):
  if i%2==0:
    count += 1
    toplam += i
# print(count , toplam, end='\n')
print(toplam/count)
```

1 saate kadar: 5 TL
1-5 saat arası: Saat başı 4 TL
5 saatten fazla: Saat başı 3 TL
Buna göre kullanıcının girdiği otoparkta kalınan saat süresine göre ödenecek miktarı bularak ekrana yazdırınız.

```python
kalış_saati = int(input('kalış saatini giriniz: '))
if kalış_saati<=1:
  print('ücretiniz:5 TL')
elif 1<kalış_saati<=5:
  print(f'Ücretiniz {kalış_saati*4}')
else:
  print(f'Ücretiniz {kalış_saati*3}')    
```
